```python
### Import File ###
import csv

""" 
Column Names: 
['Date', 'Name', 'Team', 'Season', 'Round',
    'Home Team', 'Away Team', 'Home Score', 'Away Score', 'Margin',
    'Disposals', 'Kicks', 'Marks', 'Handballs', 'Goals',
    'Behinds', 'Hitouts', 'Tackles', 'Rebounds', 'Inside 50s',
    'Clearances', 'Clangers', 'Frees For', 'Frees Against', 'Contested Pos',
    'Uncontested Pos', 'Contested Marks', 'Marks Inside 50', 'One Percenters', 'Goal Assists',
    'Brownlow Votes', 'TOG']
"""
with open("RawData.csv") as csvFile:
    csvData = csv.reader(csvFile)
    
    row1 = next(csvData)
    trnData  = {header:[] for header in row1} # Training data
    tstData = {header:[] for header in row1} # Test data
    
    for row in csvData:
        year = int(float(row[3]))
        
        # Training set
        if year < 2016: 
            for i in range(0,len(row1)):
                if i==3 or 7 <= i <= 30: # Integer Data
                    trnData[row1[i]].append( round( float( (row[i]) ) ) )
                else: # String data
                    trnData[row1[i]].append(str(row[i]))
        
        # Testing set
        else:   
            for i in range(0,len(row1)):
                if i==3 or 7 <= i <= 30: # Integer Data
                    tstData[row1[i]].append( round( float( (row[i]) ) ) )
                else: # String data
                    tstData[row1[i]].append(str(row[i]))
   
```

```python
import pandas as pd
import scipy
import matplotlib.pyplot as plt

trnDF = pd.DataFrame(trnData)
tstDF = pd.DataFrame(tstData)
```
